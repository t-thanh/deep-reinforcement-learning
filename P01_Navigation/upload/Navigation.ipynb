````python
import torch                       
from dqn_agent import Agent
from collections import deque
import matplotlib.pyplot as plt
import random

%matplotlib inline 
````

````python
reward_goal = 13
rolling_window = 100
n_episodes = 1000
number_of_trains = 10 # How many trials
````

````python
def dqn(n_episodes=n_episodes, eps_start=1.0, eps_end=0.01, eps_decay=0.995, train_no = 0, reward_goal = reward_goal):
    """Deep Q-Learning.
    
    Params
    ======
        n_episodes (int): maximum number of training episodes
        eps_start (float): starting value of epsilon, for epsilon-greedy action selection
        eps_end (float): minimum value of epsilon
        eps_decay (float): multiplicative factor (per episode) for decreasing epsilon
    """
    scores = []                                            # list containing scores from each episode
    scores_window = deque(maxlen=rolling_window)           # last 100 scores
    eps = eps_start                                        # initialize epsilon
    solved = False
    for i_episode in range(1, n_episodes+1):               # loop by episodes
        env_info = env.reset(train_mode = True)[brain_name]
        state = env_info.vector_observations[0]            # get current state 
        score = 0                                          # reset score counter
        for i in range(n_episodes):
            action = agent.act(state, eps)                 # next action from the agent 
            action = int(action)                           # change to int
            env_info = env.step(action)[brain_name]        # send action to the environment
            next_state = env_info.vector_observations[0]   # get next state
            reward = env_info.rewards[0]                   # get reward
            done = env_info.local_done[0]                  # done is true if episode has finished
            agent.step(state, action, reward, next_state, done)
            state = next_state
            score += reward
            if done:
                break 
        scores_window.append(score)       # save most recent score
        scores.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps) # decay epsilon
        
        print('\rEpisode Number: {}, Average Score: {:.2f},  Episode Score {:.2f}'. \
            format(i_episode, np.mean(scores_window), score), end="")
        
        if np.mean(scores_window) >=reward_goal and not solved: 
            solved = True
            print('\n$$$  Average reward reached reward_goal after {:d} episodes.'. format(i_episode) )
            #break
            
    torch.save(agent.qnetwork_local.state_dict(), 'checkpoint_'+str(train_no)+'.pth') # save the weights into the file 
    return scores, i_episode
````

````python
import pandas as pd
def plot_scores(scores, rolling_window=rolling_window):
    """Plot scores"""
    fig = plt.figure()
    ax = fig.add_subplot(111)
    plt.plot(np.arange(len(scores)), scores)
    plt.ylabel('Score')
    plt.xlabel('Episode #')
    rolling_mean = pd.Series(scores).rolling(rolling_window).mean()
    plt.plot(rolling_mean, linewidth=4);
    return rolling_mean
````

````python
list_fc1 = []
list_fc2 = []
list_eps_start = []
list_episodes  = []

for i in range(0, number_of_trains):
    #play with number of nodes
    fc1_nodes = random.randrange(48, 128, 8) # must be higher than 37
    fc2_nodes = random.randrange(fc1_nodes - 16 , fc1_nodes + 16, 8)

    #play with epsilon
    epsilon_start = random.randrange(985, 995, 1)/1000
    
    print('\n--------\n\rfc1_nodes: ', fc1_nodes, '|| fc2_nodes: ', fc2_nodes)
    print('\rTrain Number: ', i, '|| epsilon_start: ',epsilon_start)
    agent = Agent(state_size=state_size, action_size=action_size, seed=1, fc1_units=fc1_nodes, fc2_units=fc2_nodes)
    scores, episodes = dqn(n_episodes = n_episodes, eps_start = epsilon_start, train_no=i, reward_goal = reward_goal)
    list_fc1.append(fc1_nodes)
    list_fc2.append(fc2_nodes)
    list_eps_start.append(epsilon_start)
    list_episodes.append(episodes)
    plot_scores(scores, rolling_window=rolling_window)
````
